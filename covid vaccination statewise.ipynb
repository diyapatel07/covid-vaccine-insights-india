As there are many NULL values present in the given dataset. We need to replace those values by mean(in case of numerical data) or mode(in case of categorical data). Here, we need to work on "First Dose Administered" and "Second Dose Administered". Both of them are float, hence we will replace the Nan Values by mean(average).

#### For First Dose Administered

```python
# Average of First Dose Administered
avg_firstdose = data["First Dose Administered"].astype("float").mean(axis = 0)
print("Average of First Dose:", avg_firstdose)
```

```python
# Replacing First Dose Administered
data["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data
```

#### For Second Dose Administered

```python
# Average of Second Dose Administered
avg_seconddose = data["Second Dose Administered"].astype("float").mean(axis = 0)
print("Average of Second Dose:", avg_seconddose)
```

```python
# Replacing Second Dose Administered
data["Second Dose Administered"].fillna(value = avg_seconddose, inplace = True)
data
```

This data is ready to be used for the next questions

### Number of persons state wise vaccinated for first dose in India

```python
first_dose = data.groupby('State')[['First Dose Administered']].sum()
first_dose
```

### Number of persons state wise vaccinated for second dose in India

```python
first_dose = data.groupby('State')[['Second Dose Administered']].sum()
first_dose
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
top_states = data.groupby('State')['Total Doses Administered'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_states.values, y=top_states.index, palette='viridis')
plt.title("Top 10 States by Total Vaccine Doses Administered")
plt.xlabel("Total Doses")
plt.ylabel("State")
plt.show()
```

```python
dose_df = data.groupby('State')[['First Dose Administered', 'Second Dose Administered']].sum()
top_dose = dose_df.sort_values(by='First Dose Administered', ascending=False).head(10)

top_dose.plot(kind='bar', figsize=(14,6), colormap='coolwarm')
plt.title('First vs Second Dose Administered â€“ Top 10 States')
plt.ylabel('Doses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
